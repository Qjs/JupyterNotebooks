The math is quite beautiful:
$$z = \sum_{n=1}^N e^{2\pi if(n)} $$
In the 'daily case' we take the condition $f(n) = n/m + n^2/d + n^3/y$ where m, d, y are the month, day and year of the current day respectively. Changing this adds for some unique differences in the representation!


```python
import matplotlib.pyplot as plt
import numpy as np
import datetime
%matplotlib inline
```

```python
def f(n,m,d,y):
    return n/m + n**2/d + n**3/y
```

```python
now = datetime.datetime.now()

N = 1000
m = now.month
d = now.day
y = now.year-2000

z = array([np.exp(2*np.pi*1j*f(n,m,d,y)) for n in range(3, N+3)])
z = z.cumsum()
```

```python
plt.plot(z.real, z.imag, '-k')
plt.axes().set_aspect('equal')
```

From [UNSW](https://www.maths.unsw.edu.au/about/exponential-sums) that is mentioned in the same blog post, the 'Loch Ness Monster' graph looks amazing! (Displayed below)
$$f(n) = (\log(n))^4 $$

```python
def f2(n):
    return ((np.log(n))**4)

N = 5000
z2 = array([np.exp(2*np.pi*1j*f2(n)) for n in range(1, N)])
z2 = z2.cumsum()
plt.plot(z2.real, z2.imag, '-k')
plt.axes().set_aspect('equal')
```
